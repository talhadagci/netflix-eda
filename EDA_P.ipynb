```python
def rating(x):
    if x in ['TV-MA', 'R', 'NC-17']:
        return 'Adults'
    elif x in ['TV-14', 'PG-13']:
        return 'Teens'
    elif x in ['TV-PG', 'PG', 'TV-Y7', 'TV-Y7-FV']:
        return 'Older Kids'
    elif x in ['TV-Y', 'G', 'TV-G']:
        return 'Kids'
    else:
        return 'Unknown'
```

```python
df['country'] = df['country'].str.replace('United Kingdom', 'UK')
df['country'] = df['country'].str.replace('United States', 'USA')
df['country'] = df['country'].str.replace('South Korea', 'S. Korea')
top_10_countries = df['country'].value_counts().index[:10]
sns.countplot(data=df, x='country', order=top_10_countries,hue="country", palette='Paired')
plt.xticks(rotation=50)
plt.title('the top 10 countries')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.show()
```

```python
genres_exploded = df['listed_in'].str.split(', ').explode()
top_7_genres = genres_exploded.value_counts().head(7)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_7_genres.index, y=top_7_genres.values,hue=top_7_genres,palette='magma')
plt.title('Most Popular 7 Type', fontsize=15)
plt.ylabel('Number of Contents')
plt.xticks(rotation=45)

plt.show()
```

```python
top_10_names = df['country'].value_counts().head(10).index
df_top10 = df[df['country'].isin(top_10_names)]
gruplanmis_veri = df_top10.groupby(['country', 'type']).size().unstack()
gruplanmis_veri.plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title('Top 10 Countries: Film and TV Series Distribution')
plt.ylabel('Number')
plt.xlabel('Countries')
plt.xticks(rotation=45) 
plt.legend(title='Content Type')

plt.show()
```

```python
son_yillar = df[df['release_year'] >= 2000]
yil_veri = son_yillar.groupby(['release_year', 'type']).size().unstack()
plt.figure(figsize=(12, 6))
sns.lineplot(data=yil_veri, linewidth=2.5)
plt.title('The Speed of Film and TV Series Production by Year')
plt.xlabel('Year')
plt.ylabel('Number of Added Content')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
top_countries = df['country'].value_counts().head(5).index
top_ratings = df['rating'].value_counts().head(5).index
filtered_df = df[
    (df['country'].isin(top_countries)) & 
    (df['rating'].isin(top_ratings))
]
matris = pd.crosstab(filtered_df['country'], filtered_df['rating'])
plt.figure(figsize=(10, 6))
sns.heatmap(matris, annot=True, fmt='d', cmap='Reds') 
plt.title('Rating Distribution by Country (Density Map)')
plt.show()
```
